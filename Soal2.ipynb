```python
import cv2
import matplotlib.pyplot as plt

# Membaca gambar
img = cv2.imread('/content/image.jpg')

# Konversi BGR ke RGB agar warna benar di matplotlib
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Resize gambar ke beberapa ukuran
img_512 = cv2.resize(img, (512,512))
img_256 = cv2.resize(img, (256,256))
img_128 = cv2.resize(img, (128,128))
img_64 = cv2.resize(img, (64,64))

# Menampilkan hasil sampling
plt.figure(figsize=(10,8))

plt.subplot(2,2,1)
plt.imshow(img_512)
plt.title("512x512")

plt.subplot(2,2,2)
plt.imshow(img_256)
plt.title("256x256")

plt.subplot(2,2,3)
plt.imshow(img_128)
plt.title("128x128")

plt.subplot(2,2,4)
plt.imshow(img_64)
plt.title("64x64")

plt.show()
```
