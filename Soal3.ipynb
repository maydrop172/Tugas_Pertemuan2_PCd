```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Membaca gambar grayscale
img = cv2.imread('/content/image.jpg', cv2.IMREAD_GRAYSCALE)

# Fungsi quantization
def quantize(image, levels):

    # Menghitung interval quantization
    step = 256 // levels

    # Membulatkan nilai pixel sesuai level baru
    quantized = (image // step) * step

    return quantized

# Quantization
img_2bit = quantize(img, 4)     # 2 bit = 4 level
img_4bit = quantize(img, 16)    # 4 bit = 16 level
img_8bit = quantize(img, 256)   # 8 bit = 256 level (original)

# Menampilkan hasil
plt.figure(figsize=(10,8))

plt.subplot(2,2,1)
plt.imshow(img, cmap='gray')
plt.title("Original (8 bit - 256 level)")

plt.subplot(2,2,2)
plt.imshow(img_2bit, cmap='gray')
plt.title("2 bit (4 level)")

plt.subplot(2,2,3)
plt.imshow(img_4bit, cmap='gray')
plt.title("4 bit (16 level)")

plt.subplot(2,2,4)
plt.imshow(img_8bit, cmap='gray')
plt.title("8 bit (256 level)")

plt.show()
```
